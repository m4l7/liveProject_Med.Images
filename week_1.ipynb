```python
import numpy as np
import nibabel as nib
import glob
import ntpath 
import matplotlib.pyplot as plt

config={}
config["Data Path"] = "/home/marco/dev/PhD/live_project_manning/data/small/"
```

```python
def list_all_patients(config):

    all_files = glob.glob((config["Data Path"]+"/t1/" "*"))
    list_all_patient_id = []
    for i in all_files:

        list_all_patient_id.append(ntpath.basename(i)[:6])

    return list_all_patient_id

def load_images(pat_id, config):

    path_to_t1 = glob.glob((config["Data Path"]+ "t1/" + pat_id + "*"))
    print(path_to_t1)
    image_t1 = nib.load(path_to_t1[0])
    image_data_t1 = image_t1.get_fdata()

    path_to_t2 = glob.glob((config["Data Path"]+ "t2/" + pat_id + "*"))
    image_t2 = nib.load(path_to_t2[0])
    image_data_t2 = image_t2.get_fdata()

    return image_data_t1, image_data_t2

def plot_images(t1,t2):
    print(t1[70,:,:].shape)
    fig, axes = plt.subplots(1,2)
    axes[0].imshow(t1[:,60,:], cmap = "gray")
    axes[0].set_title("T1")
    axes[1].imshow(t2[:,60,:], cmap = "gray")
    axes[1].set_title("T2")
    plt.show()
```

```python
patients = list_all_patients(config)
print(patients)
t1,t2 = load_images(patients[5], config)
plot_images(t1,t2)
```
